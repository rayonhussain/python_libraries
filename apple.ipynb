```python
#functions
def finding_great(a=5,b=3):
    if a>b:
        print(f"{a}>{b}")
    elif b>a:
        print(f"{b}>{a}")
    else:
        print(f"{a}={b}")
```

```python
def db_cols(*args):
    for i in args:
        print(i)
        
db_cols("a","b","c","d")
```

```python
def db_cols(**kwargs):
    for i in kwargs.items(): # we give items because it identifies as dict whereas args is more like list
        print(i)
    print("Here are keys:")
    for i in kwargs:
        print(i)    
    print("Here are values:")
    for i in kwargs.values():
        print(i)
        
db_cols(c1="a",c2="b",c3="c",c4="d")
```

### Merging Data

```python
meetup_ext="/Users/rayon/apple/Archive/meetup/"
cat=pd.read_csv(meetup_ext+"categories.csv")
city=pd.read_csv(meetup_ext+"cities.csv")
g1=pd.read_csv(meetup_ext+"groups1.csv")
g2=pd.read_csv(meetup_ext+"groups2.csv")
```

```python
cat.head()
```

```python
city.head()
```

```python
#concatenation b/w df's is more like stacking on top of each other
#concat is more like union but it doesnt have to be same columns
#pd.concat(objs=[df1,df2])

groups=pd.concat(objs=[g1,g2],ignore_index=True)
groups.head()
```

```python
#merging the dataframes is like using joins
#df1.merge(df2,how="left/right/inner/outer",on="common_id",left_on,right_on)
# we can use left_on and right_on if the colm names are different in df's

groups.merge(city,how="left",left_on="city_id",right_on="id")
```

```python
groups.merge(city,how="outer",left_on="city_id",right_on="id")
```

```python
#indicator parameter gives the presence of relation(id's) in both tables or just 1
groups.merge(city,how="outer",left_on="city_id",right_on="id",indicator=True)
```
