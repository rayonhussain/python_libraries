```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=df,x='Sepal width (cm)',y='Sepal length (cm)')
```

```python
plt.figure(figsize=(13,5))
sns.violinplot(data=df,x='Sepal width (cm)')
```
