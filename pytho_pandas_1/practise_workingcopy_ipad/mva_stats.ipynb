```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
mva_df=pd.read_csv('Datasets/mva_vehiclesalescount.csv')
mva_df.head()
```

```python
print(mva_df.columns)
```

```python
#stripping the column names spaces and generalise it
mva_df.columns=mva_df.columns.str.strip().str.lower().str.replace(' ','_')
print(mva_df.columns)
```

```python
#need to change the year field into string datatype
mva_df['year']=mva_df['year'].astype(str)
print(type(mva_df['year']))
```

```python
mva_df.info()
```

```python
mva_df.groupby(by=['year','month']).mean()
```

```python
mva_df.groupby(by='year').count()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 9))  # Two subplots (top and bottom)

# Top subplot: Categorical bar chart
sns.barplot(x='year', y='new', data=mva_df, ax=axes[0], palette='pastel')
axes[0].set_title('New Car Sales per Year')

# Bottom subplot: Numerical bar chart
sns.barplot(x='year', y='used', data=mva_df, ax=axes[1], palette='dark')
axes[1].set_title('Used Car Sales per Year')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 9))  # Two subplots (top and bottom)

# Top subplot: Categorical bar chart
sns.lineplot(x='year', y='new', data=mva_df, ax=axes[0])
axes[0].set_title('New Car Sales per Year')

# Bottom subplot: Numerical bar chart
sns.lineplot(x='year', y='used', data=mva_df, ax=axes[1],color='red')
axes[1].set_title('Used Car Sales per Year')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=mva_df,x='year',y='new',color='blue')
sns.lineplot(data=mva_df,x='year',y='used',color='orange')
plt.title('Merged New and Used Car sales from 2002 to 2024')
plt.xlabel('Year')
plt.ylabel('New/Used Sales')
plt.legend(title='Legend',loc='upper right')
plt.grid(alpha=0.2)
```
